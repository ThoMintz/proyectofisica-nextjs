Diagramas de Cuerpo Libre (DCL) : TIRO EN CA√çDA LIBRE CON REBOTE

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_force_diagram(title, forces):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set_xlim(-1, 1)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal', 'box')
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])

    # Eliminar el rect√°ngulo alrededor del gr√°fico
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Dibuja el objeto (pelota)
    radio_pelota = 0.2
    circle = plt.Circle((0, 0), radio_pelota, color="blue", alpha=0.5,zorder = 2)
    ax.add_patch(circle)

    # Dibuja los ejes X e Y
    ax.axhline(0, color = 'black', linewidth = 0.5, zorder = 1)
    ax.axvline(0, color = 'black', linewidth = 0.5, zorder = 1)
     # Agregar las flechas para los ejes X e Y con etiquetas
    # Plot vectores de fuerzas
    for force, (x, y), label in forces:
        # Normaliza el vector de fuerza
        norm = np.sqrt(x**2 + y**2)
        x_normalized, y_normalized = x / norm, y / norm

        # Calcula el punto de inicio en el borde de la pelota
        start_x = x_normalized * radio_pelota
        start_y = y_normalized * radio_pelota

        # Dibuja la flecha desde el borde de la pelota
        ax.arrow(start_x, start_y, x, y, head_width=0.1, head_length=0.15, fc="red", ec="red", zorder = 2)

        # Ajusta la posici√≥n del texto basado en la direcci√≥n del vector
        text_offset = 0.2
        if y < 0:  # Fuerza hacia abajo (como la gravedad)
            ax.text(start_x-text_offset, start_y + y * 1.3 + text_offset, label, color="black", ha="center", va="top", zorder = 2)
        elif y > 0:  # Fuerza hacia arriba (como la normal)
            ax.text(start_x-text_offset, start_y + y * 1.2 -text_offset, label, color="black", ha="center", va="bottom", zorder = 2)

    plt.grid(False)
    plt.show()

# En ca√≠da libre: solo gravedad
fuerzas_cayendo = [("peso", (0, -0.8), r"$m \, \vec{g}$")]

# Durante el impacto con la mesa: gravedad y fuerza normal
fuerzas_impacto = [("peso", (0, -0.8), r"$m \, \vec{g}$"),
                   ("normal", (0, 0.8), r"$\vec{N}$")]

# Plot ambos diagramas
plot_force_diagram("Diagrama de Cuerpo Libre: Pelota en ca√≠da", fuerzas_cayendo)
plot_force_diagram("Diagrama de Cuerpo Libre: Pelota en el impacto con la mesa", fuerzas_impacto)
```

Diagramas de Cuerpo Libre (DCL): Tiro lateral con rebote

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_force_diagram(title, forces):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set_xlim(-1, 1)
    ax.set_ylim(-2, 2)
    ax.set_aspect('equal', 'box')
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])

    # Eliminar el rect√°ngulo alrededor del gr√°fico
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Dibuja el objeto (pelota)
    radio_pelota = 0.2
    circle = plt.Circle((0, 0), radio_pelota, color="blue", alpha=0.5,zorder = 2)
    ax.add_patch(circle)

    # Dibuja los ejes X e Y
    ax.axhline(0, color = 'black', linewidth = 0.5, zorder = 1)
    ax.axvline(0, color = 'black', linewidth = 0.5, zorder = 1)
    
    # Plot vectores de fuerzas
    for force, (x, y), label in forces:
        # Normaliza el vector de fuerza
        norm = np.sqrt(x**2 + y**2)
        x_normalized, y_normalized = x / norm, y / norm

        # Calcula el punto de inicio en el borde de la pelota
        start_x = x_normalized * radio_pelota
        start_y = y_normalized * radio_pelota

        # Dibuja la flecha desde el borde de la pelota
        ax.arrow(start_x, start_y, x, y, head_width=0.1, head_length=0.15, fc="red", ec="red", zorder = 2)

        # Ajusta la posici√≥n del texto basado en la direcci√≥n del vector
        text_offset = 0.2
        if y < 0:  # Fuerza hacia abajo
            if x > 0:
                ax.text(start_x + x * 1.3 + text_offset/4, start_y + y * 1.3 + text_offset, label, color="black", ha="center", va="top", zorder = 2)
            elif x <= 0:
                ax.text(start_x -text_offset, start_y + y * 1.3 + text_offset, label, color="black", ha="center", va="top", zorder = 2)
        elif y > 0:  # Fuerza hacia arriba
            if x > 0:
                ax.text(start_x + x * 1.3-text_offset, start_y + y * 1.2 +text_offset/4, label, color="black", ha="center", va="bottom", zorder = 2)
            elif (x <= 0):
                ax.text(start_x-text_offset, start_y + y * 1.2 -text_offset, label, color="black", ha="center", va="bottom", zorder = 2)
    plt.grid(False)
    plt.show()

# En ca√≠da libre
fuerzas_cayendo = [("peso", (0, -0.8), r"$m \, \vec{g}$"),
                   ("mano", (0.6,-0.3), r"$\vec{F}$")]

# Durante el impacto con la mesa: gravedad y fuerza normal
fuerzas_impacto = [("peso", (0, -0.8), r"$m \, \vec{g}$"),
                   ("normal", (0, 0.8), r"$\vec{N}$"),
                   ("mano", (0.6,0.3), r"$\vec{F}$")]

# Plot ambos diagramas
plot_force_diagram("Diagrama de Cuerpo Libre: Pelota en ca√≠da", fuerzas_cayendo)
plot_force_diagram("Diagrama de Cuerpo Libre: Pelota en el impacto con la mesa", fuerzas_impacto)
```
